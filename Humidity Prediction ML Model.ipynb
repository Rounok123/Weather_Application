### Loading the raw Dataset and showing the first 5 rows

```python
import pandas as pd
weatherHistory = pd.read_csv("weatherHistory.csv")
weatherHistory.head()
```

```python
weatherHistory.describe()
```

### Importing the immportant libraries and Data preprocessing

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
weatherFeatures = ["Temperature (C)","Apparent Temperature (C)","Wind Speed (km/h)", 
                   "Wind Bearing (degrees)","Visibility (km)","Pressure (millibars)"]

x = weatherHistory[weatherFeatures]
y = weatherHistory.Humidity

x_scaled = preprocessing.scale(x)
scaler = StandardScaler()
x_final = scaler.fit_transform(x_scaled)

## Train test spliting

x_train, x_test, y_train, y_test = train_test_split(x_final, y, test_size = 0.10, random_state = 42)
```

### Building our model

```python
from sklearn import linear_model

#Linear Regerssion training model
# model = linear_model.LinearRegression()
# model.fit(x_train, y_train)

# Ridge regression training model
model = linear_model.Ridge(alpha = 0.5)
model.fit(x_train, y_train)

# Predicting
y_predict = model.predict(x_test)
```

```python
# Testing the model accuracy

from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

print("Weights are: ", model.coef_)
print("Intercept: ", model.intercept_)
print('Mean squared error: %.3f' % mean_squared_error(y_test,  y_predict))
print('Coefficient of determination: %.3f' % r2_score(y_test, y_predict))
```

```python
### Create a Pickle file using serialization 
import pickle
pickle_out = open("model.pkl","wb")
pickle.dump(model, pickle_out)
pickle_out.close()
```
